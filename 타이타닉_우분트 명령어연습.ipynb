## 타이타닉 

### tail -n +2 titanic.csv | cut -d ',' -f 6 | sort -n | uniq -c 

'tail -n +2'      = 2열부터 시작한다.   
cut -d ',' -f 6   = ,를 기준으로 6번째 열을 자르겠다.   
sort -n           =  같은 워드로 정렬하기.   
uniq -c           =  정렬 숫자세기.
